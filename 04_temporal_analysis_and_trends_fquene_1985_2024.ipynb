# Temporal Analysis of Water Surface Area – Laguna de Fúquene (1985–2024)

This notebook analyzes temporal patterns and long-term trends in the water surface area of the Laguna de Fúquene based on the dataset generated in the previous extraction step.

## 1. Objective

The objectives of this notebook are to:
- Visualize interannual variability in water surface area
- Identify long-term trends
- Quantify the direction and magnitude of change

This notebook focuses on analysis and interpretation, not data extraction.

## 2. Libraries and data loading

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv(
    'data/processed/water_surface_area_fquene_1985_2024.csv'
)

df.head()
```

## 3. Time series visualization

The following figure shows the annual evolution of the estimated water surface area.

```python
plt.figure(figsize=(10, 5))
plt.plot(df['year'], df['area_ha'])
plt.xlabel('Year')
plt.ylabel('Water Surface Area (ha)')
plt.title('Laguna de Fúquene – Water Surface Area (1985–2024)')
plt.grid(True)
plt.show()
```

## 4. Long-term trend estimation

A linear regression is used as a first-order approximation to describe long-term changes in surface area.

```python
slope, intercept = np.polyfit(df['year'], df['area_ha'], 1)

plt.figure(figsize=(10, 5))
plt.plot(df['year'], df['area_ha'], label='Observed')
plt.plot(
    df['year'],
    intercept + slope * df['year'],
    linestyle='--',
    label='Linear trend'
)
plt.xlabel('Year')
plt.ylabel('Water Surface Area (ha)')
plt.legend()
plt.grid(True)
plt.show()

slope
```

## 5. Interpretation

The estimated trend represents an average long-term change and should be interpreted cautiously. Interannual variability, sensor limitations, and methodological assumptions influence the magnitude of the observed trend.

## 6. Limitations

- Linear trends may oversimplify complex dynamics
- Classification uncertainty propagates into area estimates
- Results depend on ROI definition and threshold selection

These limitations are inherent to long-term satellite-based analyses and do not invalidate the results.
