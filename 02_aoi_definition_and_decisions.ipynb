# AOI Definition and Spatial Decisions

This notebook focuses on the definition of the Area of Interest (AOI) and the spatial decisions
that condition all subsequent analyses. The choice of spatial boundaries is treated as a
decision-making step rather than a purely technical operation.


## 1. Decision context

In environmental analysis based on satellite data, spatial definitions directly influence
quantitative results, uncertainty levels, and the type of conclusions that can be drawn.

From a decision-making perspective, defining an AOI implies selecting which territories,
processes, and stakeholders are included or excluded from the analysis.


## 2. Why AOI definition is a decision

The AOI is not an objective or fixed entity. Different AOI definitions may lead to different
estimates of environmental variables such as surface water extent, vegetation cover, or land
use dynamics.

These differences can influence policy recommendations, resource allocation, and monitoring
strategies.


## 3. Criteria for AOI selection

The AOI used in this project is defined based on the following criteria:

- Relevance to the environmental process under study
- Availability and consistency of satellite observations
- Administrative or natural boundaries meaningful for decision-making
- Reproducibility and transparency of spatial definitions


## 4. Spatial data sources

The AOI geometry is derived from vector data sources that allow explicit control over
spatial boundaries. Using vector-defined AOIs ensures that spatial decisions are
documented and reproducible.

## 5. Implications of AOI geometry

Small changes in AOI geometry may result in:

- Inclusion or exclusion of transitional land covers
- Changes in pixel counts and area estimates
- Different sensitivity to mixed pixels

These effects are especially relevant when using medium-resolution satellite imagery.


## 6. Spatial resolution and scale considerations

The interaction between AOI size, shape, and sensor spatial resolution is a critical
element of the analysis. AOIs that are too small may be dominated by edge effects,
while overly large AOIs may dilute localized processes.


## 7. Uncertainty and spatial decisions

Spatial uncertainty is introduced not only by sensor characteristics, but also by
the subjective choices involved in AOI delineation.

Recognizing this uncertainty is essential when interpreting results for decision-making
purposes.
