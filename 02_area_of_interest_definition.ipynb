# Area of Interest (AOI): Laguna de Fúquene

This notebook defines the spatial extent used for the satellite and climate analysis.
A consistent Area of Interest (AOI) is critical to ensure reproducibility and comparability across datasets and time.


## Geographic Context

The Laguna de Fúquene is located in the Ubaté Valley, on the border between the departments of Cundinamarca and Boyacá, Colombia.

The lagoon lies at approximately 2,540 meters above sea level and is part of a high-altitude Andean wetland system.


## Spatial Definition Strategy

Instead of using administrative boundaries, the AOI is defined as a bounding polygon
that fully encloses the historical maximum extent of the lagoon.

This approach avoids underestimating surface water changes during high-water periods
and ensures consistent satellite sampling across time.


## AOI Bounding Box Coordinates (WGS84)

- Minimum Longitude: -73.92
- Maximum Longitude: -73.78
- Minimum Latitude: 5.40
- Maximum Latitude: 5.50


## AOI GeoJSON Definition


```python
import geopandas as gpd
from shapely.geometry import shape

geojson = {
    "type": "FeatureCollection",
    "features": [{
        "type": "Feature",
        "properties": {"name": "Laguna de Fuquene AOI"},
        "geometry": {
            "type": "Polygon",
            "coordinates": [[
                [-73.92, 5.40],
                [-73.78, 5.40],
                [-73.78, 5.50],
                [-73.92, 5.50],
                [-73.92, 5.40]
            ]]
        }
    }]
}

geometry = [shape(feature["geometry"]) for feature in geojson["features"]]

aoi_gdf = gpd.GeoDataFrame(
    geojson["features"],
    geometry=geometry,
    crs="EPSG:4326"
)

aoi_gdf

```

```python
aoi_gdf.plot(edgecolor="black", facecolor="none", figsize=(5,5))
```
