# Long-Term Satellite Monitoring of the Laguna de Fúquene

## Objective
This notebook defines the analytical scope, data sources, and key metrics used to study long-term changes in the surface water area of the Laguna de Fúquene (Colombia) using satellite-derived datasets and climate data.


## Problem Context

The Laguna de Fúquene is one of the most important wetland systems in the Andean region of Colombia. Over recent decades, the lagoon has experienced significant ecological and hydrological stress, driven by a combination of climatic variability and anthropogenic pressure.

Understanding long-term changes in the surface water extent of the lagoon is critical for environmental monitoring, water resource management, and policy decision-making.

This project focuses on quantifying and analyzing the long-term stability of the lagoon's surface water area and exploring its relationship with precipitation patterns.


## Analytical Questions

1. How has the surface water area of the Laguna de Fúquene evolved over the last four decades?
2. Is there a detectable long-term trend or structural change in water surface stability?
3. How does precipitation variability relate to observed changes in surface water extent?
4. Are satellite-derived trends consistent across independent data sources?


## Data Sources

### 1. Satellite-Derived Surface Water Data
- Source: JRC Global Surface Water Dataset (European Commission)
- Temporal coverage: 1984–present
- Spatial resolution: 30 meters
- Description: Provides historical information on surface water presence derived from Landsat imagery.

### 2. Climate Data
- Source: NASA POWER / CHIRPS API
- Variable: Precipitation
- Temporal aggregation: Monthly
- Spatial reference: Coordinates corresponding to the Laguna de Fúquene region


## Scope and Methodological Choices

- The analysis focuses exclusively on the open water surface of the Laguna de Fúquene.
- A long-term temporal window (1985–2024) is used to capture structural trends and reduce short-term noise.
- Surface water area is selected as the primary indicator of hydrological stability.
- Climate variables are analyzed as explanatory factors rather than direct causal drivers.


## Limitations

- Satellite-derived classifications may contain uncertainty due to cloud cover and sensor limitations.
- The analysis does not explicitly model human interventions such as water extraction or land-use change.
- Results should be interpreted as indicative of long-term trends rather than precise hydrological measurements.
