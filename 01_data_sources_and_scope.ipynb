# Decision context, data sources and project scope

This notebook defines the decision context, data sources, analytical scope, and initial methodological choices of the project **Decision Making with Satellite Data: Evolution of the surface water extent of Laguna de Fúquene (1985–2024)**.

## 1. Introduction

Laguna de Fúquene is one of the most important water bodies in the Altiplano Cundiboyacense and has undergone significant changes in its surface extent over time.

The objective of this project is to analyze the temporal evolution of surface water extent using Earth Observation (EO) data, prioritizing a reproducible workflow and a critical interpretation of results.

Rather than producing a definitive estimate of the lagoon’s area, this analysis focuses on understanding how satellite data, their inherent limitations, and methodological decisions influence the estimates obtained.

From a decision-making perspective, this project examines how satellite-derived indicators and their uncertainty condition the type of conclusions that can be drawn for environmental monitoring and territorial management.

## 2. Earth Observation (EO)

Earth Observation (EO) refers to the use of satellite, aerial, or ground-based sensors to monitor physical processes on the Earth’s surface.

In this project, EO is used to:
- Detect surface water through spectral information
- Analyze spatial and temporal changes
- Evaluate the quality and reliability of historical satellite data

EO does not provide exact values, but rather estimates that depend on sensor characteristics, spatial resolution, temporal coverage, and processing decisions.

In this context, EO is treated as a decision-support tool rather than a source of absolute truth, emphasizing its role in comparative analysis and long-term trend evaluation.

## 3. Satellite data sources

The analysis is primarily based on data from the Landsat mission, using imagery from:

- Landsat 5
- Landsat 7
- Landsat 8

These satellites provide a continuous temporal record starting in 1984, enabling the analysis of nearly 40 years of surface water evolution in a consistent manner.

Main characteristics of the data:
- Spatial resolution: 30 meters
- Global coverage
- Free access and scientific validation
- Availability through Google Earth Engine

## 4. Sensor selection: justification

Although more recent sensors such as Sentinel-2 offer higher spatial resolution, this project prioritizes temporal consistency over spatial detail.

The main reasons for selecting Landsat are:
- Continuous temporal coverage since 1984
- Comparability between years without changing sensors
- Reduction of inconsistencies introduced by instrumental differences
- Suitability for long-term trend analysis

Sentinel-2 was explored preliminarily but excluded from the main analysis due to its limited availability prior to 2015.

## 5. Study area and polygon definition

The study area corresponds to Laguna de Fúquene, Colombia.

A polygon (Region of Interest, ROI) was defined to cover the full known extent of the water body. During early stages of the project, it was identified that initial polygon versions only partially covered the lagoon, resulting in significant underestimation of surface area.

The review and correction of the polygon constituted a critical step to ensure spatial coherence of the analysis.

This step illustrates how spatial definitions directly affect quantitative results and, consequently, the decisions that may be informed by them.

## 6. Project scope

This project has an exploratory and analytical scope.

The analysis aims to:
- Evaluate temporal trends in surface water extent
- Analyze how different methodological decisions affect results
- Identify limitations inherent to satellite-derived data

The project does not aim to:
- Replicate exact official figures
- Estimate water volume
- Correct or interpolate missing data
- Attribute direct causality to climatic or anthropogenic drivers

## 7. Discarded data and approaches

Several approaches were evaluated during the development of the project and ultimately discarded:

- Monthly analysis, due to high levels of missing data and cloud-related noise
- Interpolation of missing observations, to avoid introducing unverifiable assumptions
- Full combination of multiple sensors across the entire temporal series

These decisions form part of the learning process and the critical evaluation of the analysis.

## 8. Known limitations

From the outset, the following limitations are acknowledged:

- 30 m spatial resolution limits the detail of shoreline delineation
- Frequent cloud cover in the region
- Complex morphological changes of the water body
- Dependence on threshold-based water classification

These limitations are not considered errors, but fundamental elements for interpreting the results.
