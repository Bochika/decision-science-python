# Fuentes de datos y alcance del proyecto

Este notebook define las fuentes de datos utilizadas, el alcance del análisis y las decisiones metodológicas iniciales del proyecto **Análisis de la evolución del espejo de agua de la Laguna de Fúquene (1985–2024)**.

## 1. Introducción

La Laguna de Fúquene es uno de los cuerpos de agua más importantes del altiplano cundiboyacense y ha experimentado transformaciones significativas en su extensión superficial a lo largo del tiempo.

El objetivo de este proyecto es analizar la evolución temporal del espejo de agua utilizando datos de Observación de la Tierra (EO), priorizando un enfoque reproducible y una interpretación crítica de los resultados.

Este análisis no busca producir una cifra definitiva del área de la laguna, sino evaluar cómo los datos satelitales, sus limitaciones y las decisiones metodológicas influyen en las estimaciones obtenidas.

## 2. Observación de la Tierra (EO)

La Observación de la Tierra (Earth Observation, EO) consiste en el uso de sensores satelitales, aéreos o terrestres para monitorear fenómenos físicos del planeta.

En este proyecto, EO se utiliza para:
- Detectar agua superficial mediante información espectral
- Analizar cambios espaciales y temporales
- Evaluar la calidad y confiabilidad de datos satelitales históricos

La EO no proporciona valores exactos, sino estimaciones dependientes del sensor, la resolución espacial, la cobertura temporal y las decisiones de procesamiento.

## 3. Fuentes de datos satelitales

El análisis se basa principalmente en datos de la misión Landsat, utilizando imágenes de:

- Landsat 5
- Landsat 7
- Landsat 8

Estos satélites proporcionan una serie temporal continua desde 1984, lo que permite analizar casi 40 años de evolución del espejo de agua de manera consistente.

Características principales de los datos:
- Resolución espacial: 30 metros
- Cobertura global
- Acceso gratuito y validación científica
- Disponibilidad a través de Google Earth Engine

## 4. Selección del sensor: justificación

Aunque sensores más recientes como Sentinel-2 ofrecen mejor resolución espacial, este proyecto prioriza la consistencia temporal sobre el detalle espacial.

Las razones principales para seleccionar Landsat son:
- Continuidad temporal desde 1984
- Comparabilidad entre años sin cambiar de sensor
- Reducción de inconsistencias introducidas por diferencias instrumentales
- Adecuación para análisis de tendencias de largo plazo

Sentinel-2 fue evaluado de forma exploratoria, pero descartado para el análisis principal debido a su disponibilidad limitada a partir de 2015.

## 5. Área de estudio y definición del polígono

El área de estudio corresponde a la Laguna de Fúquene, Colombia.

Se definió un polígono (ROI) que cubre la extensión completa conocida del cuerpo de agua. Durante el desarrollo del proyecto se identificó que versiones iniciales del polígono solo cubrían parcialmente la laguna, lo que resultaba en subestimaciones significativas del área.

La revisión y corrección del polígono fue un paso clave para garantizar la coherencia espacial del análisis.

## 6. Alcance del proyecto

Este proyecto tiene un alcance exploratorio y analítico.

El análisis busca:
- Evaluar tendencias temporales en el espejo de agua
- Analizar cómo distintas decisiones metodológicas afectan los resultados
- Identificar limitaciones inherentes a los datos satelitales

El proyecto no busca:
- Replicar cifras oficiales exactas
- Estimar volumen de agua
- Corregir o interpolar datos faltantes
- Atribuir causalidad directa a factores climáticos o antrópicos

## 7. Datos y enfoques descartados

Durante el desarrollo del proyecto se evaluaron varios enfoques que finalmente fueron descartados:

- Análisis mensual, debido al alto número de datos faltantes y ruido por nubosidad
- Interpolación de datos faltantes, para evitar introducir supuestos no verificables
- Uso combinado de múltiples sensores en toda la serie temporal

Estas decisiones forman parte del aprendizaje y del análisis crítico del proyecto.

## 8. Limitaciones conocidas

Desde el inicio del proyecto se reconocen las siguientes limitaciones:

- Resolución espacial de 30 m limita el detalle de los bordes
- Nubosidad frecuente en la región
- Cambios morfológicos complejos del cuerpo de agua
- Dependencia de umbrales para la clasificación de agua

Estas limitaciones no se consideran errores, sino elementos fundamentales para interpretar los resultados.
